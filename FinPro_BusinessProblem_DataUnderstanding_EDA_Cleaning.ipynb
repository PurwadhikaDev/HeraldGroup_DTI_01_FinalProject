## **1. Introduction**
### **1.1 Context**
A Portuguese bank is conducting telemarketing campaigns to encourage customers to make term deposits. The bank wants to improve its campaign efficiency and increase deposit conversion rates, especially given the importance of deposits for bank stability highlighted by the 2008 financial crisis. The bank has data from previous campaigns, including customer demographics, contact details, campaign outcomes, and relevant economic indicators. They want to use this data to predict which customers are more likely to agree to a term deposit, allowing them to target their efforts more effectively and reduce wasted resources.

Target:

0: Customer did not make a term deposit

1: Customer made a term deposit

### **1.2 Telemarketing Campaign Overview**
- Total Customers Contacted: 41,188
- Total Subscriptions Achieved: 4,640
- Average telephone cost: 5 EURO (Based on industry standards, the average cost per call can range from 2.70 EURO to 5.60 EURO ([Average telephone cost](https://www.liveagent.com/customer-support-glossary/cost-per-call/)), we set it at 5 EURO)
- Minimum Deposit Required: 100 EURO (According to typical banking practices, the minimum balance required to open a savings account is often around 100 EURO ([Minimum deposit](https://www.nerdwallet.com/article/banking/savings/typical-minimum-balances-to-open-a-savings-account)))


**Lets calculate current conversion rate**
$$
\text{Conversion Rate} = \left( \frac{\text{Total Subscriptions Achieved}}{\text{Total Customers Contacted}} \right) \times 100 = \left( \frac{4,640}{41,188} \right) \times 100 \approx 11.3\%
$$
<center>(Conversion rate measures the proportion of customers who actually make a term deposit compared to the number of customers contacted through the campaign)</center>


Current conversion rate is 11.3%, but there is still significant room for improvement, especially when compared to top industry performers who achieve conversion rates of up to 23%. [Top industry performers](https://www.ruleranalytics.com/blog/reporting/financial-services-marketing-statistics/#:~:text=Marketers%20in%20the%20financial%20services,at%20a%20rate%20of%2023%25)


### **1.3 Problem Statement**
Telemarketing campaigns for term deposits can be over budget if the Portuguese bank targets all customers without proper screening. The bank wants to improve its campaign efficiency by identifying which customers are more likely to agree to a term deposit, allowing them to focus their efforts more effectively.

With the current conversion rate standing at a mere 11.3%, which is alarmingly low, the bank risks significant financial losses if it continues to contact all potential customers indiscriminately. This inefficiency could lead to unnecessary expenses and reduced overall campaign effectiveness, potentially impacting the bank's ability to secure the deposits crucial for its stability and operations. So based by the problem, our concerns are:

Problem Statement for Machine Learning
- How to predict customer conversion probability for term deposits so we can optimize our telemarketing campaign efficiency and increase our conversion rate beyond top industry performance that achive conversion rate up to 23%?

Problem Statement for Analytics
- What customers characteristics and interactions should we focus on to maximize conversion rates?

### **1.4 Objectives**
The main objectives of our project are :
- Develop a predictive model to forecast the probability of customers opening a term deposit, enabling the financial team to optimize budget allocation and reduce unnecessary expenses in telemarketing campaigns.
- Identify key factors influencing customer decisions to open term deposits, providing the marketing team with actionable insights for targeted customer engagement and improved conversion rates.

### **1.4 Analytical Approach**
- Step 1, we will conduct a comprehensive analysis of the existing dataset to identify the key characteristics of customers who have shown interest in term deposits. This process will help us determine which features have the most significant impact on a customer's decision to open a term deposit.
- Step 2, based on the insights gained from the data analysis, we will construct a machine learning classification model using the most influential features. This model will enable the bank to predict the probability of a customer's willingness to open a term deposit, thereby optimizing their telemarketing campaign strategy and improving conversion rates.

### **1.5 Metrics**
Given our objectives of Data Analysis and Machine Learning Model development, our project will utilize different metrics tailored to each purpose. For data analysis, our primary metric will be the conversion rate, while for the machine learning model, our focus will be on precision. By enhancing the precision of our model, we aim to improve the conversion rate, thereby making the telemarketing campaign more effective in targeting customers who are likely to open term deposits.

- **Type I Error (False Positive)**: This error occurs when the model incorrectly predicts that a customer will open a term deposit when they actually do not. Each false positive leads to an unnecessary telemarketing cost of 5 EURO per customer. For instance, if there are 1000 false positives, the total ineffective telemarketing cost would amount to 5000 EURO. Reducing this error is crucial for minimizing wasted marketing expenditures.

- **Type II Error (False Negative)** : This error happens when the model incorrectly predicts that a customer will not open a deposit when they actually do. While this error does not directly impact the marketing budget, as these customers are not contacted, it represents a missed opportunity to convert potential depositors. However, since the focus is on cost efficiency, Type II errors are less critical than Type I errors in this context. Therefore, we value this type II error as 0.

By improving the precision of our model, we aim to reduce Type I errors, which will, in turn, increase the conversion rate. Higher precision means that the telemarketing campaign will more effectively target customers who are indeed likely to open deposits. This approach minimizes wasted resources on unlikely customers and ensures that the marketing budget is allocated more efficiently, leading to better campaign results and overall higher conversion rates.

```python
from scipy.stats import normaltest
alpha = 0.05
index = 1

dist_info = []
for feature_name in numerical_columns:
    stat, p_value = normaltest(df_source[feature_name].dropna())

    if p_value > alpha:
        distribution = "Normally Distributed"
    else:
        distribution = "Not Normally Distributed"

    skewness = df_source[feature_name].skew()

    if skewness > 0.3:
        skewness_type = "Right Skew"
    elif skewness < 0.3:
        skewness_type = "Left Skew"
    else:
        skewness_type = "Normal"

    dist_info.append([feature_name, stat, p_value, distribution, skewness, skewness_type])

distribution_info = pd.DataFrame(dist_info, columns=["Feature",
                                                    "D'Agostino-Pearson Statistic",
                                                    "P-value",
                                                    "Distribution",
                                                    "Skewness",
                                                    "Skewness Type"])
distribution_info
```

```python
def eda_plot(col):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=[12, 5])
    proportion = df_source2[col].value_counts()
    rate = df_source2[[col, 'deposit']].groupby(col)['deposit'].mean().sort_values()

    # Menggunakan palet warna dengan gradasi dibalik
    sns.barplot(x=proportion.index, y=proportion.values, ax=ax1, palette='crest_r')
    ax1.set_title(f'{col} Proportion', fontsize=12)
    ax1.set_xlabel(None)
    ax1.set_ylabel('Frequency', fontsize=8)
    plt.sca(ax1)
    plt.xticks(rotation=45, fontsize=8)

    sns.barplot(x=rate.index, y=rate.values, ax=ax2, palette='crest_r')
    ax2.set_title(f'{col} Deposit Rate', fontsize=12)
    ax2.set_xlabel(None)
    ax2.set_ylabel('Deposit Rate', fontsize=8)
    ax2.axhline(y=0.5, color='red', linestyle='--')
    plt.sca(ax2)
    plt.xticks(rotation=45, fontsize=8)
    plt.tight_layout()
    plt.show()

    yes = df_source2[[col, 'deposit']].groupby(col)['deposit'].sum()
    all = 5 * df_source2[col].value_counts()
    waste = (all - yes) / 1000

    waste = waste.reset_index()
    waste.columns = [col, 'waste']

    fig = px.treemap(waste, path=[col], values='waste',
                     title=f'Treemap of {col} Budget Waste')

    fig.update_traces(
        textinfo='label+value',
        textfont_size=14,
        texttemplate='%{label}<br>EUR %{value}K',
        selector=dict(type='treemap')
    )

    fig.show()
```

```python
def eda_line(df_line, col):
    color_palette = sns.color_palette("crest_r")

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=[14, 6])

    sns.histplot(df_line[col], bins=12, kde=True, ax=ax1, color=color_palette[0])
    ax1.set_title(f'{col} Proportion', fontsize=14)
    ax1.set_xlabel(None)
    ax1.set_ylabel('Frequency', fontsize=10)
    ax1.grid(True, linestyle='--', alpha=0.7)
    plt.sca(ax1)
    plt.yticks(fontsize=8)
    plt.xticks(fontsize=8)
    
    sns.lineplot(data=df_line, x=col, y='deposit', ax=ax2, ci=None, color=color_palette[1])
    ax2.axhline(y=0.5, color='red', linestyle='--')
    ax2.set_title(f'{col} Deposit Rate', fontsize=14)
    ax2.set_xlabel(None)
    ax2.set_ylabel('Deposit Rate', fontsize=10)
    ax2.grid(True, linestyle='--', alpha=0.7)
    plt.sca(ax2)
    plt.yticks(fontsize=8)
    plt.xticks(fontsize=8)
    
    fig.tight_layout()
    plt.show()

    df_line['waste'] = [5 if val == 0 else 0 for val in df_line['deposit']]
    line = df_line.groupby(col).agg({'waste': 'sum'}).reset_index()

    plt.figure(figsize=[14, 6])
    sns.lineplot(data=line, x=col, y='waste', ci=None, color=color_palette[2])
    plt.title(f'Total Budget Waste Based on {col}', fontsize=14)
    plt.xlabel(None)
    plt.ylabel('Total Budget Waste', fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.xticks(fontsize=8)
    plt.yticks(fontsize=8)
    plt.tight_layout()
    plt.show()
```

```python
def plot_categorical_threshold(data, categorical_columns, threshold=0.5):
    num_cols = len(categorical_columns)
    num_rows = (num_cols + 2) // 3
    
    fig, axes = plt.subplots(num_rows, 3, figsize=(15, 4 * num_rows), constrained_layout=True)
    
    axes = axes.flatten()
    
    for i, col in enumerate(categorical_columns):
        if col in data.columns:
            deposit_rate = data[[col, 'deposit']].groupby(col)['deposit'].mean().sort_values()
            
            sns.barplot(x=deposit_rate.index, y=deposit_rate.values, palette='crest_r', ax=axes[i])
            axes[i].set_title(f'Deposit Rate by {col}')
            axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=45, fontsize=8)
            axes[i].set_xlabel(None)
            axes[i].axhline(y=threshold, color='red', linestyle='--', label=f'Threshold {threshold}')
            axes[i].legend()
        else:
            # Hide empty subplots
            axes[i].set_visible(False)
    
    # Hide any remaining empty subplots
    for j in range(num_cols, len(axes)):
        axes[j].set_visible(False)
    
    plt.show()
```

### **4.3 Fill in Missing Values**

```python
# show descriptive statistics before processing
stats_before = df_source.describe(include='object')
stats_before
```

```python
# Define the orderings for the relevant categorical columns
orderings = {
    'marital': ['single', 'married', 'divorced', 'unknown'],
    'education': ['illiterate', 'basic 4y', 'basic 6y', 'basic 9y', 'high school', 'professional course', 'university degree', 'unknown'],
    'job': ['unemployed', 'retired', 'student', 'housemaid', 'admin', 'services', 'blue-collar', 'technician', 'management', 'self-employed', 'entrepreneur', 'unknown'],
    'default': ['no', 'yes', 'unknown'],
    'housing': ['no', 'yes', 'unknown'],
    'loan': ['no', 'yes', 'unknown'],
    'month': ['mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec'],
    'contact': ['telephone', 'cellular'],
    'day_of_week': ['mon', 'tue', 'wed', 'thu', 'fri'],
    'poutcome': ['nonexistent', 'failure', 'success']
}
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

df_unknown = df_source.copy()
col_unknown = [col for col in df_unknown.columns if df_unknown[col].isin(['unknown']).any()]

cat_cols = [col for col in df_unknown.columns if col in orderings]
for col in cat_cols:
    encoder = OrdinalEncoder(categories=[orderings[col]])
    df_unknown[col] = encoder.fit_transform(df_unknown[[col]])


for col in col_unknown:
    nan_encoder = len(orderings[col]) - 1
    df_unknown[col] = df_unknown[col].replace(nan_encoder, np.nan)


unrelated_feature = ['emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed', 'deposit']
related_feature = [col for col in df_unknown.columns if col not in unrelated_feature]

imputer = IterativeImputer(max_iter=10, random_state=42, initial_strategy='most_frequent')
df_unknown_imputed = imputer.fit_transform(df_unknown[related_feature])
df_unknown[related_feature] = df_unknown_imputed.astype(int)

for col in cat_cols:
    inverse_mapping = {i: value for i, value in enumerate(orderings[col])}
    df_unknown[col] = df_unknown[col].map(inverse_mapping)
```

```python
# show descriptive statistics before processing
stats_after = df_unknown.describe(include='object')
display(stats_before,stats_after)
```

> **Insight**

In general, after applying Iterative Imputer to the dataset, the characteristics of each variable do not change significantly. This is expected, as data imputation should not substantially alter the distribution of the data. The only noticeable change is the removal of the 'unknown' category, which previously represented missing values.

### **4.4 Check Outliers**

```python
df_unknown.describe().T
```

```python
numerical_columns = df_unknown.select_dtypes(include=['float', 'int']).columns

def detect_outliers(df_unknown, column):
    Q1 = df_unknown[column].quantile(0.25)
    Q3 = df_unknown[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_unknown.loc[(df_unknown[column] < lower_bound) | (df_unknown[column] > upper_bound)]

    print(f"Column: {column}")
    print(f"Lower Bound: {lower_bound}")
    print(f"Upper Bound: {upper_bound}")
    print(f"The number of outliers in '{column}' is {len(outliers)} or {(len(outliers)/len(df_unknown)*100):.2f}% of the total data ({len(df_unknown)} rows).")
    print("-" * 90)

for column in numerical_columns:
    detect_outliers(df_unknown, column)
```

```python
outliers_col = ['age', 'campaign', 'pdays', 'previous']
outlier_bounds = {}

for col in outliers_col:
    Q1 = df_unknown[col].quantile(0.25)
    Q3 = df_unknown[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier_bounds[col] = {"lower_bound": lower_bound, "upper_bound": upper_bound}
```

```python
plt.figure(figsize=(18, 12))
for i, col in enumerate(outliers_col):
    value_counts_deposit = df_unknown[(df_unknown[col] > outlier_bounds[col]["upper_bound"]) & (df_unknown["deposit"] == 1)][col].value_counts()
    value_counts = df_unknown[df_unknown[col] > outlier_bounds[col]["upper_bound"]][col].value_counts()

    percentage = np.round(value_counts_deposit / value_counts, 2)
    percentage = percentage.fillna(0)

    plt.subplot(2, 2, i + 1)
    percentage.plot(kind="bar", color='skyblue')
    plt.axhline(y=0.1177, color='red', linestyle='--', linewidth=2)
    plt.xlabel(col)
    plt.ylabel(f"Probability of deposit given {col} value")
    plt.title(f"Probability of Deposit Among Outliers in {col} Feature")

plt.tight_layout()
plt.show()
```

> **Insight:**
- Age: Outliers for ages above 85 tend to have a higher probability of deposit, suggesting that older individuals may be more likely to make a deposit.
- Campaign: Outliers with a very high number of contacts tend to have a lower probability of deposit, indicating that excessive marketing efforts can negatively impact deposit decisions.
- Pdays: Outliers show significant variation in deposit probability, implying that the time between the last marketing contact and the current one significantly influences deposit decisions.
- Previous: Outliers with higher values for previous contacts show a higher probability of deposit, indicating that clients who have been contacted multiple times before are more likely to make a deposit.
